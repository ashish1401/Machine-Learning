```python
#write all the features

#everything is a feature except the price column , coz we need to find that out
```

```python
X = car.drop(columns = 'Price')
y = car.Price
```

```python
from sklearn.model_selection import train_test_split
#split dataset into training and testing dataset
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
X_train , X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)
```

```python
ohe = OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_), ['name' , 'company' , 'fuel_type']),remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
#comes in handy when while deploying we are not aware of the pre processors
pipe = make_pipeline(column_trans,lr)
```

```python
#pipelne created first generates encoded numeric values of all features using OneHotEncoder
```

```python
pipe.fit(X_train,y_train)
#pehle column transformer lagayega , then unpe encoder then will give the processed data
```

```python
ypred = pipe.predict(X_test)
```

```python
r2_score(y_test,ypred)
```

```python
#now since the data set is divided to training and testing datasets randomly ,
#we run a loop till we find a given training set and testing set with maximum r2_score
```

```python
store = []
for i in range(1000):
  X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2 , random_state=i)
  lr = LinearRegression()
  pipe = make_pipeline(column_trans,lr)
  pipe.fit(X_train,y_train)
  ypred = pipe.predict(X_test)
  store.append(r2_score(y_test,ypred))
```

```python
np.argmax(store)
```

```python
store[np.argmax(store)] #accuracy
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size = 0.2, random_state = np.argmax(store))
lr = LinearRegression()
pipe = make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
ypred = pipe.predict(X_test)
print(r2_score(y_test,ypred))
```
